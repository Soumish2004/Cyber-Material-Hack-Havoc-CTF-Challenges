```python

with open("cRYPTO.txt", "r") as file:
    contents = file.read()
```

```python
arr = contents.split(",")
```

```python
arr = [chr(int(i,2)) for i in arr]
arr
```

```python
ans = []
t=''
for i in arr:
    if i == ',':
        ans.append(t)
        t=''
        continue
    t+=i
```

```python
ans
```

```python
ans = [chr(int(i,2)) for i in ans]
```

```python
ans
```

```python
ans1 = []
t=''
for i in ans:
    if i == ' ':
        ans1.append(t)
        t=''
        continue
    t+=i
```

```python
ans1 = [chr(int(i,2)) for i in ans1]
```
